```python
import pyproj
import pandas as pd
import numpy as np
```

EPSG:32611 - WGS 84 / UTM zone 11N
EPSG:4326 - World Geodetic System 1984

```python
gcp=pd.read_csv(r'C:\Users\taohuang\Desktop\gcp_epsg32611_2021_los1.csv')
gcp
```

```python
wgs84 = pyproj.Proj(projparams = 'epsg:4326')
utm = pyproj.Proj(projparams = 'epsg:32611')

#x1, y1 = -11705274.6374,4826473.6922
#pyproj.transform(gcp, wgs84, x1, y1)
```

```python
for row in gcp.index:
    #print( gcp.loc[row,['x','y']])
    x=gcp.loc[row,['x']]
    y=gcp.loc[row,['y']]
    x,y=np.array([x,y])
  #  print(pyproj.transform(utm, wgs84, x, y)[0].astype("float"))
    gcp.loc[row,['lat']]=pyproj.transform(utm, wgs84, x, y)[0].astype("float")
    gcp.loc[row,['lon']]=pyproj.transform(utm, wgs84, x, y)[1].astype("float")
```
